```python
import pandas as pd
titanic=pd.read_excel('titanic.xls')
titanic_data=titanic.drop(['body','home.dest'], axis=1)
titanic_data
```

```python
titanic_data['age_cat']=pd.cut(titanic_data['age'], bins=[0,3,7,15,30,60,100], include_lowest=True, labels=['baby','children','teenage','young','adult','old'])
titanic_data.groupby('age_cat', observed=True).mean(numeric_only=True)
#열의 값을 특정 기준으로 나눌때; cut 함수를 사용/bins ; 경계 구간 설정
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt

data=titanic_data
plt.figure(figsize=(14,4))
plt.subplot(131)
sns.barplot(x='pclass',y='survived',data=titanic_data)
plt.subplot(132)
sns.barplot(x='age_cat',y='survived',data=titanic_data)
plt.subplot(133)
sns.barplot(x='sex',y='survived',data=titanic_data)
plt.subplots_adjust(top=1, bottom=0.1, left=0.10,right=1,hspace=0.5,wspace=0.5)
plt.show()
```

```python
titanic_data.groupby(['pclass','sex']).mean(numeric_only=True)
```

```python
boat_survivors=titanic_data[titanic_data['boat'].notna()] #boat값이 NaN이 아닌행만 추출
nboat_survivors=titanic_data[titanic_data['boat'].isna()] #NaN인 행만 추출.
print(boat_survivors.groupby('sex')['survived'].mean())
print(nboat_survivors.groupby('sex')['survived'].mean())
```

```python
import pandas as pd
from scipy import stats

ds=[60,62,64,65,68,69]
weight=pd.Series(ds)
ds.append(120)
weight_heavy=pd.Series(ds)
weight_heavy
```

```python
print(weight.mean()) #평균은 이상치에 영향을 받음
print(weight_heavy.mean())
```

```python
print(weight.median()) # 중앙값; 데이터를 크기순으로 정렬했을 때 가장 중앙에 위치하는 값.
print(weight_heavy.median()) # 중앙값은 이상치에 크게 영향을 받지 않음.
```

```python
print(stats.trim_mean(weight,0.2)) #절사평균 ; 하위/상위 0.2%를 제외 후 남은 값들로 평균 계산.
print(stats.trim_mean(weight_heavy,0.2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("iris.csv")

#화면 분할 정의
fig, axes=plt.subplots(nrows=2,ncols=2)

#각 분할 영역에 그래프 작성하기
df['Petal_Length'].plot.hist(ax=axes[0,0])
df['Petal_Width'].plot.box(ax=axes[0,1])

fd=df['Species'].value_counts()
fd.plot.pie(ax=axes[1,0])
fd.plot.barh(ax=axes[1,1])

#통합 그래프에 제목 지정
fig.suptitle('Multiple Graph Example', fontsize=14)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("iris.csv")

fig, axes=plt.subplots(nrows=2,ncols=2)

df['Petal_Length'].plot.hist(ax=axes[0,0])
df['Petal_Width'].plot.box(ax=axes[0,1])

fd=df['Species'].value_counts()
fd.plot.pie(ax=axes[1,0])
fd.plot.barh(ax=axes[1,1])

fig.suptitle('Multiple Graph Example', fontsize=14)
# tight_layout()메서드로 subplot의 제목, 축라벨, 눈금 등이 안겹치도록
fig.tight_layout()
plt.show()
```
