```python
import pandas as pd

x=[1,2,3,4,5]
y=[0,2,4,6,8]
df=pd.DataFrame()
df['x_value']=x
df['y_value']=y

df
```

```python
df['x*y']=df['x_value']*df['y_value']
df
```

```python
df2=df.copy()
df2['number']=['zero','one','two','three','four']
df2.set_index('number',append=True)
```

```python
df2
```

```python
df3 =df2.set_index('number',append=True) #df.set_index('열 이름',append=True), drop처럼 원본 안바뀜.
df3
```

```python
df4=df3.reset_index(drop=True) #reset_index;인덱스 초기화
df4
```

```python
dates=pd.date_range('20250901',periods=6) #pd.date_range; 날짜로 인덱스 만들기.
dates
```

```python
dates=pd.date_range('20250901',periods=6,freq='YS') #YS(Year Start)
dates
```
