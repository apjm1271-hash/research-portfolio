```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

beers=[5,2,9,8,3,7,3,5,3,5]
bal=[0.1,0.03,0.19,0.12,0.04,0.0095,0.07,0.06,0.02,0.05]
dict={'beers':beers,'bal':bal}
df=pd.DataFrame(dict)
df
```

```python
df.plot.scatter(x='beers',y='bal',title="Beers-blood alchhol level")
plt.show()
```

```python
df.plot.scatter(x='beers',y='bal',title="Beers-blood alchhol level")
m,b=np.polyfit(beers,bal,1)
plt.plot(beers,m*np.array(beers)+b)
plt.show()
```

```python
df['beers'].corr(df['bal']) # 
```

```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
import matplotlib.pyplot as plt

df=pd.read_csv('BostonHousing.csv')
df=df[['crim','rm','dis','tax','medv']]
titles=['crime','rooms','distance','tax','medv']
```

```python
grp=pd.Series(['M' for i in range(len(df))])
grp.loc[df['medv']>=25.0]='H'
grp.loc[df['medv']<=17.0]='L'
df['grp']=grp
df
```

```python
new_odr=['H',"M","L"]
new_dtype=CategoricalDtype(categories=new_odr,ordered=True)
df.grp=df['grp'].astype(new_dtype)
df['grp'].dtype
```

```python
print(df.shape)
df.head()
```

```python
print(df.dtypes)
df['grp'].value_counts(sort=False)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)

for i in range(5):
  df[df.columns[i]].plot.hist(ax=axes[i//3,i%3],ylabel='',xlabel=titles[i])

fig.suptitle('H',fontsize=14)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
fig.tight_layout()
#fig.subplots_adjust(hspace=0.5,wspace=0.3)
#   ?


for i in range(5):
  df[df.columns[i]].plot.box(ax=axes[i//3,i%3],ylabel='',xlabel=titles[i])

fig.suptitle('B',fontsize=14)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)

for i in range(5):
  df[df.columns[i]].plot.box(ax=axes[i//3,i%3],ylabel='',xlabel=titles[i])

fig.suptitle('B',fontsize=14)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
fig.subplots_adjust(hspace=0.5,wspace=0.3)

for i in range(5):
  df.boxplot(column=df.columns[i],by='grp',grid='False',
             ax=axes[i//3,i%3],fontsize=4)

fig.suptitle('Boxplot by group',fontsize=14)
plt.show()
```
