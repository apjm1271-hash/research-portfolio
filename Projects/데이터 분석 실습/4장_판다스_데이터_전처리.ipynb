```python
#with ; 열고 닫기가 자동.
with open('file.txt', 'r', newline='') as filereader:
  for line in filereader:
    print("{}".format(line.strip())) #변수.strip() #여백제거
```

```python
#여러 파일 읽기 : os, glob module.
#glob.glob(os.path.join("*.txt") 현재 폴터에 있는 모든 .txt파일을 찾아서 리스트로 반환. 이후 for file in : glob이 반환한 파일 이름 리스트를 하나씩 file 변숭 저장.
#os module txt로끝나는 모든 파일 경로를 찾음. 이후 join으로 합침/확장자를 합침 os;파일 관리자.
import os
import glob

for file in glob.glob(os.path.join("*.txt")):
  with open(file, 'r', newline='', endcoding="UTF8") as filereader:
    for row in filereader:
      print("{}".format(row.strip()))

#glob은 하나씩 열어서 조사. glob ; 파일 검색기
#UTF8 encoding 방식

strip();좌우/lstrip;/rstrip 각각 왼쪽/오른쪽 공백&지정문자 제거 + 가운데 공백은 안없어짐
```

```python
# pivot_table ;기존 표데이터로 새롭게 표 생성 /행, 열 값을 뭘로할지 지정가능. aggfunc ; sum,mean etc/pivot_table의 기본 aggfunc : mean
t_df.pivot_table(index="pclass",columns='sex',values='survived')
```

```python
t_df.groupby('survived').mean(numeric_only=True)
```

```python
groupby 데이터프레임의 특정 열을 기준으로 묶어서 표시 groupby().count()/sum, mean도 가능 ;mean() 괄호에 아무것ㅅ도 안넣으면 오류듬
mean-숫자형에만 사용가능 그래서 mean(numeric_only=True) 이거 해줘야됨 즉 기본값은 numeric_only=False로 돼있음.
```
